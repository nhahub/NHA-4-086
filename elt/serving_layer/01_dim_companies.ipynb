```python
from pyspark.sql import functions as F

# ==========================================================
# Configuration
# ==========================================================

SILVER_TABLE = "finance_intelligence_hub.silver.companies"
GOLD_VIEW = "finance_intelligence_hub.gold.dim_companies"

print("="*80)
print("Companies Gold View")
print("="*80)

# ==========================================================
# Create / Replace Gold View
# A VIEW (not a table) so it always reflects the latest silver data with
# zero duplicate storage. CREATE OR REPLACE is always safe to re-run —
# it redefines the view only, and never touches silver's actual rows.
# ==========================================================

spark.sql(f"""
CREATE OR REPLACE VIEW {GOLD_VIEW} AS
SELECT
    ticker,
    company_name,
    category,
    price,
    change,
    change_percent,
    volume,
    avg_volume_3m,
    market_cap,
    pe_ratio,
    week52_change_percent,
    week52_low,
    week52_high
FROM {SILVER_TABLE}
""")

print(f"Gold view {GOLD_VIEW} created/refreshed successfully.")

# ==========================================================
# Quick sanity check
# ==========================================================

row_count = spark.sql(f"SELECT COUNT(*) FROM {GOLD_VIEW}").first()[0]
print(f"Gold view row count: {row_count:,}")

spark.sql(f"DESCRIBE {GOLD_VIEW}").show(truncate=False)
```
