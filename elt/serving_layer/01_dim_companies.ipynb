```sql

CREATE OR REPLACE VIEW finance_intelligence_hub.gold.dim_companies AS

SELECT

    -- ---- Identity ----
    si.ticker,
    COALESCE(c.company_name, si.company_name)  AS company_name,
    si.short_name,
    si.exchange,
    si.currency,
    si.country,
    si.sector,
    si.industry,
    si.quote_type,
    si.shares_outstanding,

    -- ---- Business / classification (companies only) ----
    c.category,

    -- ---- Market metrics not available elsewhere ----
    -- (price/change/volume live in silver.stock_prices instead)
    c.market_cap                                AS market_cap_usd,
    c.pe_ratio,
    c.week52_change_percent,
    c.week52_low,
    c.week52_high,

    -- ---- Raw reference market cap from stocks_info ----
    -- NOTE: not currency-normalized, kept for cross-check only
    si.market_cap                                AS market_cap_reference_raw,

    -- ---- Audit ----
    GREATEST(
        COALESCE(si.silver_loaded_at, TIMESTAMP('1900-01-01')),
        COALESCE(c.silver_loaded_at,  TIMESTAMP('1900-01-01'))
    )                                             AS gold_loaded_at

FROM finance_intelligence_hub.silver.stocks_info si

JOIN finance_intelligence_hub.silver.companies c
    ON si.ticker = c.ticker
```
